```python
import numpy as np
import pickle
import random

from agent import Agent

np.random.seed(0)
random.seed(0)
```

```python
start_pos = [(0, 0), (8, 8), (8, 0), (0, 8)]
batch_size = 2500
total_itrs = 10000

aggregated_episodes = []

for i in range(total_itrs):
    idx = (i // batch_size) % len(start_pos)
    start = start_pos[idx]
    new_agent = Agent(start = start, goal=(4, 4), x_max=9, y_max=9, reward=1000)
    
    episodes = new_agent.train(200)
    aggregated_episodes.append(episodes)

    if (i + 1) % batch_size == 0:
        filename = f'/Users/User/Downloads/TLSAV/Datasets/Agent_Batch_{start}.pkl'
        with open(filename, 'wb') as file:
            pickle.dump(aggregated_episodes, file)
        aggregated_episodes = []

    print(f"Currently Training Agent {i} Starting At {start}")
```
